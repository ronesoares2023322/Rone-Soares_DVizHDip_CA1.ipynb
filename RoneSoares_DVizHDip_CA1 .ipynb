>### What are the top 5 games by global sales? 

#### I've created a bar chart showing 5 games with the highest global sales. I used the nlargest() method to select the 5 rows with the largest values in column global_sales.

* I used pandas because it offers the function that makes it easy to select the main values.

* I also chose Seaborn to create a bar chart because of its simplicity and ability to produce compelling visualizations with a few lines of code.

* I chose the Bar chart because it's common for comparing categories to each other, like games. In this case, it also allows me a quick understanding of the differences in the overall sales of the top 5 games.

* To plot this graph I opted for a 12x8 figure. To ensure that the chart is displayed in a clear and readable manner. The style of the chart I used was Seaborn's "Whitegrid".

* To add white gridlines to the background of the chart, making it easier to read.

* I also chose a magma color palette for the colors of the Bars, as it provides a range of distinct and quite attractive colors.

* I've also added the label of the axes, I've added labels to the X and Y axes to make it clear what each axis represents.

* I've also adjusted the size and font of the text on the labels and Bar labels to make sure they're legible and visually pleasing.

* In the values part, I've displayed the values of global sales in each bar. To provide specific information about the data, these values are displayed in a readable and clear manner.

```python
#Top 5 games by global sales
top5_games = data[['Name', 'Global_Sales']].nlargest(5, 'Global_Sales')

# Plot settings
plt.figure(figsize= (12,8))
sns.set_style('whitegrid')


# Creating bar plot and labels on bars
top5_games_plot=sns.barplot(data=top5_games, x= 'Global_Sales', y= 'Name', palette='magma')
for index, value in enumerate(top5_games['Global_Sales']):
    plt.text(value + 0.1, index, f'{value}M' , va= 'center', color= 'black')
    
# Chart settings
plt.title('Top 5 Games by Global Sales', fontsize= 16)
plt.xlabel('Global Sales (millions)', fontsize= 14)
plt.ylabel('Game Name', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

#Displaying the values on bars
for i, v in enumerate(top5_games['Global_Sales']):
    plt.text(v + 0.1, i, f'{v}', va='center', color='black', fontweight='bold')
plt.tight_layout()
plt.show()




```

> ## Is there a correlation between the na_sales and jp_sales for the years 2010-2014?
* This code performs several tasks related to the correlation analysis between sales in North America and sales in Japan for the years 2010 to 2014.

* I used pandas' "corr()" method to calculate the correlation between the NA_SALES and JP_SALES columns of the filtered dataframe, which contains only the data from the years 2010 to 2014. I did this to examine the relationship between sales in North America and Japan during this period.

* I used Seaborn to create a scatter plot as a density heatmap. This allows you to visualize the joint distribution of sales in North America and Japan and identify visual patterns in the relationship between them.

* 2- I chose a scatter plot because it's effective for visualizing the relationship between two continuous variables, such as sales in North America and Japan, it allows you to identify patterns and trends in the data.

* The density heatmap has been overlaid on the scatter plot to highlight the regions where the data points are densest. This provides a visual representation of the joint distribution of sales.

* 3- To plot the graph, I opted for a figure. 10x6 size to ensure that the graphic is displayed clearly and legibly.

* The chart style I used was Seaborn's Whitegrid to add white grid lines to the background of the chart, making it easier to read.

* I also used a transparency of 0.7 and the size of the markers as being 100 to improve the visualization of the data points.

* In the colors I opted for one from the 'Viridis' palette to provide a range of distinct and pleasing colors to the graphic. These colors make it easy to distinguish between the different areas of the density heatmap.

* I've added labels to the X and Y axes to make it clear what each axis represents.

* I put a descriptive title for the chart and a legend for the color bar of the density heatmap to help with the interpretation of the data.


```python
# Filtering the data and calculating the correlation
filtered_data = data[(data['Year_of_Release']>= 2010) & (data['Year_of_Release'] <= 2014)] 
correlation = filtered_data['NA_Sales'].corr(filtered_data['JP_Sales'])
print(f" The correlation between NA_Sales and JP_Sales for the years 2010 to 2014 is:{correlation:.2f}")
```

```python
# Creating scatter plot with density heatmap
plt.figure(figsize= (10,6))
sns.set_style("whitegrid")
sns.scatterplot(data= filtered_data, x= 'NA_Sales', y='JP_Sales', alpha= 0.7, s=100, palette= 'viridis')
sns.kdeplot(data=filtered_data, x= 'NA_Sales', y='JP_Sales', cmap='viridis', shade=True, alpha=0.5, cbar=True, cbar_kws={'label':'Density'})
plt.title('Correlation between NA_Sales and JP_Sales(2010-2014)', fontsize=16)
plt.xlabel ('NA_Sales(millions)', fontsize=14 )
plt.ylabel ('JP_Sales(millions)', fontsize=14 )
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

>## Do older games (2005 and earlier) have a higher MEAN eu_sales than newer games (after 2005)?  


* To answer this question, I created a Horizontal Bar chart using Plotly Express to show the average European sales for games before and after 2005.

* 1- Plotly Express offers a simple and expressive way to create interactive visualizations. I chose the "px.bar()" method. To create a horizontal bar chart, as it is effective for comparing values between two categories.

* The data was prepared in a dataframe using pandas, where I averaged European sales for the games before and after 2005 and organized them into a dataframe called Sales_data.

* 2- To plot the graph, I used a horizontal bar chart, which is suitable for comparing values between 2 categories, in this case, game sales before and after 2005.

* I chose a divergent "blugrn" color palette, which helps highlight the differences between the sales averages before and after 2005, with shades of blue for smaller values and shades of green for larger values.

* 3- I have added a descriptive title to the Chart to provide context on the data presented.

* I've added clear labels for the X and Y axes to indicate what each axis represents.

* I chose a divergent "blugrn" color palette, to highlight the difference between the values of the European sales averages before and after 2005.

* I've added text labels to display the average sales values in each bar, making it easier to understand the differences between the groups.

```python
import plotly_express as px
old_games=data[data["Year_of_Release"] <=2005]
new_games=data[data["Year_of_Release"]>2005]

mean_eu_sales_old=old_games[ 'EU_Sales'].mean()
mean_eu_sales_new=new_games['EU_Sales'].mean()

sales_data= pd.DataFrame({
    'Year Range': ['Before 2005', 'After 2005'],
    'Mean EU Sales': [mean_eu_sales_old,mean_eu_sales_new]
}) 

fig = px.bar(sales_data,
             y='Year Range',
             x= 'Mean EU Sales', 
             text= 'Mean EU Sales',
             orientation='h',
             title='Average European Sales for Games Before and After 2005', 
             labels={'Year Range': 'Year Range', 'Mean EU Sales': 'Mean European Sales'}, 
             color='Mean EU Sales', 
             color_continuous_scale='blugrn') 
fig.update_traces(texttemplate='%{text:2f}', textposition= 'outside') 
fig.show()
             
```

>### I created a line graph using Express plotly to show the evolution of European sales over time, before and after 2005.

*  I used pandas' "groupby" method to calculate the average European sales for the games before and after 2005, grouped by year of release.

* I used the pd.concat() method to combine the resulting data frames into a single Dataframe called sales_data, with an additional column indicating whether the games are old or new.

* Plotly Express provides an easy way to create interactive visualizations, here I used px.line() to plot the line graph.

*  I chose a line graph because it's more suitable to show the evolution of a variable over time, here it allows you to visualize how European game sales changed before and after 2005.

* Segmenting the rows by old games and recent games makes it easy to compare sales trends over time for each group.

*  I added a descriptive title to the chart to provide context about the data presented.

* Include clear labels for the XEY axes to indicate what each region represents.

* I used different colors to distinguish the lines, representing game sales before and after 2005.

* Define specific axis ranges to ensure that all data is included in the visualization and for a better understanding of trends over time.

```python
import plotly_express as px
old_games=data[data["Year_of_Release"] <=2005]
new_games=data[data["Year_of_Release"]>2005]

mean_eu_sales_old=old_games.groupby('Year_of_Release')['EU_Sales'].mean().reset_index()
mean_eu_sales_new=new_games.groupby('Year_of_Release')['EU_Sales'].mean().reset_index()

sales_data= pd.concat([ mean_eu_sales_old.assign(Games='Old Games'), mean_eu_sales_new.assign(Games='New Games')])

fig = px.line(sales_data, 
              x= 'Year_of_Release', 
              y= 'EU_Sales', 
              color= 'Games',
              title='Evolution of European Sales Before and After 2005',
              labels= {'Year_of_Release': 'Year_of_Release','EU Sales': 'European Sales' },
                range_x= [1980,2020], 
                range_y= [0,sales_data['EU_Sales'].max() *1.1]) 
fig.show()
```

### Is there a relationship between Critic's Scores and overall Game Sales?
>Based on the correlation between the critics' score and the overall sales of the games observed in the graph, we can affirm that there is a positive but moderate relationship between these variables. The correlation between critics' score and overall sales is 0.21, which suggests that, in general, games with higher scores tend to have higher overall sales. However, it is important to point out that this relationship is not extremely strong, indicating that other factors also influence global sales of the games.

* I chose a scatter plot. Because it allows you to visualize the relationship between 2 continuous variables, such as the score of the critics, users, and the overall sales.

* I used different colors for the data points, according to the source of the score (critics or users) to make them easier to identify.

* I've added legends to indicate the correlation calculated for each dataset. Which provides additional information about the strength and direction of the relationship between the variables.

* I've put a clear title and axis labels to make the visualization understandable and informative.

```python
data['User_Score']= pd.to_numeric(data['User_Score'], errors= 'coerce')
filtered_data= data.dropna(subset= ['Critic_Score', 'User_Score', 'Global_Sales'])

correlation_sales_critic= filtered_data['Critic_Score'].corr(filtered_data['Global_Sales'])
correlation_sales_user= filtered_data['User_Score'].corr(filtered_data['Global_Sales'])

plt.figure(figsize=(12,6))
sns.scatterplot(x=filtered_data['Critic_Score'], y=filtered_data['Global_Sales'], color='b', label=f'Critic Score Correlation:{correlation_sales_critic:.2f}')
sns.scatterplot(x=filtered_data['User_Score'], y=filtered_data['Global_Sales'], color='r', label=f'User Score Correlation:{correlation_sales_user:.2f}')
plt.title('Correlation betwen Critic Scores/Users Score and Global Sales', fontsize=16)
plt.xlabel('Score', fontsize=14)
plt.ylabel('Global Sales', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
