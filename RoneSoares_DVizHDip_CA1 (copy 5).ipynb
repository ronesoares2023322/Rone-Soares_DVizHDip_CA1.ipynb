```python
data['Year_of_Release'].fillna(0, inplace=True)
data['Year_of_Release'] = data['Year_of_Release'].astype(int)
```

```python
data['User_Score'].replace('tbd', np.nan, inplace=True)


data['Critic_Score'] = data['Critic_Score'].astype(float)
data['Critic_Count'] = data['Critic_Count'].astype(float)
data['User_Score'] = data['User_Score'].astype(float)
data['User_Count'] = data['User_Count'].astype(float)


Critic_Score_mean = data['Critic_Score'].mean()
Critic_Count_mean = data['Critic_Count'].mean()
User_Score_mean = data['User_Score'].mean()
User_Count_mean = data['User_Count'].mean()

data['Critic_Score'].fillna(Critic_Score_mean, inplace=True)
data['Critic_Count'].fillna(Critic_Count_mean, inplace=True)
data['User_Score'].fillna(User_Score_mean, inplace=True)
data['User_Count'].fillna(User_Count_mean, inplace=True)
```

```python
#Top 5 games by global sales
top5_games = data[['Name', 'Global_Sales']].nlargest(5, 'Global_Sales')

# Plot settings
plt.figure(figsize= (12,8))
sns.set_style('whitegrid')


# Creating bar plot and labels on bars
top5_games_plot=sns.barplot(data=top5_games, x= 'Global_Sales', y= 'Name', palette='magma')
for index, value in enumerate(top5_games['Global_Sales']):
    plt.text(value + 0.1, index, f'{value}M' , va= 'center', color= 'black')
    
# Chart settings
plt.title('Top 5 Games by Global Sales', fontsize= 16)
plt.xlabel('Global Sales (millions)', fontsize= 14)
plt.ylabel('Game Name', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

#Displaying the values on bars
for i, v in enumerate(top5_games['Global_Sales']):
    plt.text(v + 0.1, i, f'{v}', va='center', color='black', fontweight='bold')
plt.tight_layout()
plt.show()




```

```python
# Filtering the data and calculating the correlation
filtered_data = data[(data['Year_of_Release']>= 2010) & (data['Year_of_Release'] <= 2014)] 
correlation = filtered_data['NA_Sales'].corr(filtered_data['JP_Sales'])
print(f" The correlation between NA_Sales and JP_Sales for the years 2010 to 2014 is:{correlation:.2f}")
```

```python
# Creating scatter plot with density heatmap
plt.figure(figsize= (10,6))
sns.set_style("whitegrid")
sns.scatterplot(data= filtered_data, x= 'NA_Sales', y='JP_Sales', alpha= 0.7, s=100, palette= 'viridis')
sns.kdeplot(data=filtered_data, x= 'NA_Sales', y='JP_Sales', cmap='viridis', shade=True, alpha=0.5, cbar=True, cbar_kws={'label':'Density'})
plt.title('Correlation between NA_Sales and JP_Sales(2010-2014)', fontsize=16)
plt.xlabel ('NA_Sales(millions)', fontsize=14 )
plt.ylabel ('JP_Sales(millions)', fontsize=14 )
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Filtering the top 4 most polpular games genres
top_genres= data['Genre'].value_counts().head(4).index
filtered_data = data[data['Genre'].isin(top_genres)]
plt.figure(figsize=(10,6))
sns.set_style('whitegrid')

sns.countplot(data=filtered_data, x= 'Genre', order= top_genres, palette= 'mako', width= 0.5)
plt.title('Distribuition of the Top 4 Most Popular Games Genres', fontsize=16)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Number of Games', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import plotly_express as px
old_games=data[data["Year_of_Release"] <=2005]
new_games=data[data["Year_of_Release"]>2005]

mean_eu_sales_old=old_games[ 'EU_Sales'].mean()
mean_eu_sales_new=new_games['EU_Sales'].mean()

sales_data= pd.DataFrame({
    'Year Range': ['Before 2005', 'After 2005'],
    'Mean EU Sales': [mean_eu_sales_old,mean_eu_sales_new]
}) 

fig = px.bar(sales_data,
             y='Year Range',
             x= 'Mean EU Sales', 
             text= 'Mean EU Sales',
             orientation='h',
             title='Average European Sales for Games Before and After 2005', 
             labels={'Year Range': 'Year Range', 'Mean EU Sales': 'Mean European Sales'}, 
             color='Mean EU Sales', 
             color_continuous_scale='blugrn') 
fig.update_traces(texttemplate='%{text:2f}', textposition= 'outside') 
fig.show()
             
```

```python
import plotly_express as px
old_games=data[data["Year_of_Release"] <=2005]
new_games=data[data["Year_of_Release"]>2005]

mean_eu_sales_old=old_games.groupby('Year_of_Release')['EU_Sales'].mean().reset_index()
mean_eu_sales_new=new_games.groupby('Year_of_Release')['EU_Sales'].mean().reset_index()

sales_data= pd.concat([ mean_eu_sales_old.assign(Games='Old Games'), mean_eu_sales_new.assign(Games='New Games')])

fig = px.line(sales_data, 
              x= 'Year_of_Release', 
              y= 'EU_Sales', 
              color= 'Games',
              title='Evolution of European Sales Before and After 2005',
              labels= {'Year_of_Release': 'Year_of_Release','EU Sales': 'European Sales' },
                range_x= [1980,2020], 
                range_y= [0,sales_data['EU_Sales'].max() *1.1]) 
fig.show()
```
